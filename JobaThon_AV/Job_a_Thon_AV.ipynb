```python
cat=['City_Code','Accomodation_Type','Reco_Insurance_Type','Is_Spouse','Health Indicator','Holding_Policy_Type','Reco_Policy_Cat']
num=['Region_Code','Upper_Age','Lower_Age','Holding_Policy_Duration','Reco_Policy_Premium']
```

```python
train_c=train[cat]
train_n=train[num]
```

```python
train_c['Reco_Policy_Cat']=train_c['Reco_Policy_Cat'].astype('str')
train_n['Holding_Policy_Duration']=train_n['Holding_Policy_Duration'].astype('float')
```

```python
test_c=test[cat]
test_n=test[num]
```

```python
test_c['Reco_Policy_Cat']=test_c['Reco_Policy_Cat'].astype('str')
test_n['Holding_Policy_Duration']=test_n['Holding_Policy_Duration'].astype('float')
```

```python
pd.get_dummies(train_c,drop_first=True)
```

```python
train_c1=pd.get_dummies(train_c,drop_first=True)
test_c1=pd.get_dummies(test_c,drop_first=True)
```

```python
train_final=pd.concat([train_n,train_c1],axis=1)
test_final=pd.concat([test_n,test_c1],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(class_weight="balanced")    
lr.fit(train_final,train.Response)
preds=lr.predict_proba(test_final)
```

```python
train_final=(train_final-train_final.mean())/train_final.std()
test_final=(test_final-test_final.mean())/test_final.std()

import xgboost as xgb
from sklearn.metrics import roc_auc_score
from sklearn.metrics import mean_squared_error
xg_reg = xgb.XGBRegressor(objective ='binary:logistic', colsample_bytree = 0.6, learning_rate = 0.1,
                max_depth = 6, alpha = 0.1, n_estimators = 1000) #colsample_bytree = 0.4, learning_rate = 0.075
#xg_reg.fit(X_train,y_train)
xg_reg.fit(train_final,train.Response)
print(roc_auc_score(train.Response, xg_reg.predict(train_final)))
preds = xg_reg.predict(test_final)
```

```python
train_final=(train_final-train_final.mean())/train_final.std()
test_final=(test_final-test_final.mean())/test_final.std()

train_final1=pd.concat([train_final,train_final[train.Response==1].reset_index().iloc[:,1:],train_final[train.Response==1].reset_index().iloc[:,1:]],axis=0)
resp=pd.concat([train.Response,pd.DataFrame(np.ones(len(train_final1)-len(train.Response)))],axis=0)#.iloc[:,0].value_counts()
resp.columns=(['Response'])
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([    
    keras.layers.Dense(512,kernel_initializer='normal', activation='relu',input_shape=[len(train_final.columns)]),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(2)
])

model.summary()

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['mse','mae','accuracy'])
model.fit(train_final1, resp.Response, epochs=35)

#test_loss, test_acc = model.evaluate(X_test,  y_test, verbose=2)

#print('\nTest accuracy:', test_acc)

probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])

predictions = probability_model.predict(test_final)
print(predictions)
preds=predictions[:,1]
```

```python
submission=pd.concat([test.ID,pd.DataFrame(preds)],axis=1)
submission.columns=(["ID","Response"])
```
